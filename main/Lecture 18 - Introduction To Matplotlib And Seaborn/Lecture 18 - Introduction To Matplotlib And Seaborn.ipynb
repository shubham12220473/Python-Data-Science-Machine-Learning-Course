# **Introduction to Matplotlib and Seaborn**


**Definition:**

* **Matplotlib:** Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides an

object-oriented API for embedding plots into applications using general-purpose GUI toolkits.

**Seaborn:** Seabom is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Use Case in Real Life:

* **Data Exploration:** Create various plots to visualize data distributions and relationships between variables during the exploratory data analysis (EDA) phase.

* **Statistical Analysis:** Use visualizations to understand statistical properties of datasets, such as distribution plots, histograms, and pair plots to identify correlations and patterns.

* **Business Reporting:** Generate business reports with visualizations that provide insights into sales performance, customer behavior, and market trends.


**Creating a simple Line Plot with Matplotlib**

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt

#Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]

#Creating a Line Plot
plt.plot(x,y,ls = "-",color='r',lw="2",marker="*",mfc="b",ms="9") #ls - linestyle , lw - line width , mfc - marker color , ms - marker size
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
import numpy as np
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

**Creating a Scatter plot with matplotlib**

```python
#Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]

#Creating a Scatter Plot
plt.scatter(x,y)
plt.xlabel("Sales")
plt.ylabel("Prices")
plt.title("Simple Scatter Plot")
plt.show()
```

**Creating a Simple Bar Plot with matplotlib**

```python
#Data
products = ['PC','Laptop','Tablet','TV']
sales = [100,200,300,400]

#Creating a Bar Plot
plt.bar(products,sales,color="orange",width=0.5)
plt.xlabel("Products")
plt.ylabel("Sales (in 1000s)")
plt.title("Simple Bar Plot")
plt.show()
```

**Creating a Simple Histogram with Matplotlib**

```python
#Data
data = [1,2,2,3,3,3,4,4,4,4]

#Creating a Simple Histogram
plt.hist(data,bins=4)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')
plt.show()
```

**Creating SubPLots with Matplotlib**





```python
#Data
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]

#Creating subplots
fig, axs = plt.subplots(2)
axs[0].plot(x,y1)
axs[0].set_title("First Plot")
axs[1].plot(x,y2)
axs[1].set_title("Second Plot")

#Displaying the plot
plt.tight_layout()
plt.show()
```

**Adding Annotations with matplotlib**

```python
#Data
x=[1,2,3,4,5]
y=[2,3,5,7,11]

#Creating a plot with annotations
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Annotations')
plt.annotate('Peak',xy=(5,11), xytext=(4,10),
             arrowprops=dict(facecolor='black',shrink=0.05))

plt.show()
```

**Creating a simple Line plot with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data
x1 = [1,2,3,4,5]
y1 = [2,3,5,7,11]

#Creating a Line Plot
sns.lineplot(x=x1,y=y1)
plt.xlabel('X=axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot with Seaborn')
plt.show()
```

**Creating a simple Scatter Plot with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data
x=[1,2,3,4,5]
y=[2,3,5,7,11]

#Creating a Simple Scatter Plot
sns.scatterplot(x=x,y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.show()
```

**Simple Bar Plot with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data
Products=['PC','TV','AC']
Values = [4,7,8]

#Creating a Simple Bar Plot
sns.barplot(x=Products,y=Values)
plt.xlabel('Products')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

**Creating a simple histogram with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data
x=[1,2,2,3,3,3,4,4,4,4]

sns.histplot(x,bins=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Histogram')
plt.show()
```

**Creating a Pair Plot with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris
iris_data = sns.load_dataset("iris")

#Creating a pair plot
sns.pairplot(iris_data,hue="species")
plt.title("Pair Plot with Seaborn")
plt.show()
```

**Creating a HeatMap with Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10,12)
sns.heatmap(data)
plt.title('HeatMap with Seaborn')
plt.show()
```
